```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as gp
```

```python
data = pd.read_excel('datos argentina.xls')
```

```python
indicadores= data['Indicator Name']
dict(indicadores ) #los datos para construir la piramida poblacional van de indice 25 a 64 inclusive

gdp = data.iloc[16:21]

inflation = data.iloc[23:25]
gdp
```

```python
fig, pan1 = plt.subplots( sharex=True, constrained_layout=True, figsize=(15,6))

pan2=pan1.twinx()

aÃ±os=inflation.columns[1:-1]
inflation_x_year = inflation.iloc[0,1:-1]
inflation_cumulative = inflation.iloc[1,1:-1]

pan1.plot(aÃ±os, inflation_x_year, color = 'red', label = 'inflacion por aÃ±o')
pan2.plot(aÃ±os, inflation_cumulative, color = 'green', label='inflacion acumulada')

pan1.set_ylabel('inflacion por aÃ±o')
pan2.set_ylabel('inflacion acumulada')
pan1.set_xlabel('AÃ±o')
for tick in pan1.get_xticklabels():
    tick.set_rotation(45)

plt.show()
```

```python
fig, pan1 = plt.subplots( sharex=True, constrained_layout=True, figsize=(12,6))

pan2=pan1.twinx()
pan3=pan1.twinx()

aÃ±os=gdp.columns[1:-1]
gdp_x_year = gdp.iloc[0,1:-1]
gdppp = gdp.iloc[2,1:-1]
gdp_growth = gdp.iloc[1,1:-1]

pan1.plot(aÃ±os, gdp_x_year, color = 'red', label = 'PBI')
pan2.plot(aÃ±os, gdppp, color = 'green', label='PBI per capita')
pan3.plot(aÃ±os, gdp_growth, color = 'blue', label='crecimieento del PBI (%)')

pan1.set_ylabel('PBI (Millones de U$S)')
pan2.set_ylabel('PBI per capita (U$S)')
pan3.set_ylabel('crecimiento del PBI (%)')
pan1.set_xlabel('AÃ±o')
fig.legend(bbox_to_anchor=(1.0, 1.0), loc='upper left')
for tick in pan1.get_xticklabels():
    tick.set_rotation(90)
    
pan3.spines["right"].set_position(("axes", 1.1))

plt.show()
```
